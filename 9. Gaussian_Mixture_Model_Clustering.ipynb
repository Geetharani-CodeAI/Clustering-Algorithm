```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("Mall_Customers.csv")      
dataset
```

```python
x = dataset.iloc[:,[3,4]].values
x
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_scaler = sc.fit_transform(x)
```

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=4, random_state=42)
y_gmm = gmm.fit_predict(x_scaler)
```

```python
# Plot clusters
plt.figure(figsize=(8,6))
plt.scatter(x_scaler[:,0], x_scaler[:,1], c=y_gmm, cmap='coolwarm', s=50, alpha=0.7)

# Highlight Cluster centers
plt.scatter(gmm.means_[:,0], gmm.means_[:,1], c='red', marker='X', s=200, label='Cluster Centers')
plt.title("GMM Clustering (Mall Customers)")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```
